```
from pandas import read_csv
dataframe = read_csv('IndiaNormalized.csv', header=0,engine='python',usecols=range(1,2000))
dataframe.shape
```

```
from keras.layers import TimeDistributed
import numpy
import matplotlib.pyplot as plt
from pandas import read_csv
import math
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.layers import LSTM
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
```

```
print(dataframe.shape)
#print(dataframe.iloc[1,:])
dataset = dataframe.values
#print(dataset[2,:])
dataset = dataset.astype('float32')
#print(dataset.shape)
#print(dataset[0])
scaler = MinMaxScaler(feature_range=(0, 1))
dataset = scaler.fit_transform(dataset)
#print(dataset[1,:])
dataset = dataset.reshape(no_days,no_data_per_day,features)# no. of days, days data, features
#print(dataset.shape)
#print(dataset[0,:,:])
```

```
print(dataset.shape)
#train_size = int(len(dataset) * 0.67)
#test_size = len(dataset) - train_size
train_size=727
test_size = len(dataset) - train_size
train, test = dataset[0:train_size,:], dataset[train_size:len(dataset),:]
print(train.shape, test.shape)
#print (train)
#print(train[0])
#print(test[0])
```

```
def create_dataset(dataset, look_back=1):
	dataX, dataY = [], []
	for i in range(len(dataset)-look_back-1):
		a = dataset[i:(i+look_back),:,:]
		dataX.append(a)
		dataY.append(dataset[i + look_back,:,:])
	return numpy.array(dataX), numpy.array(dataY)
```

```
import numpy
# reshape into X=t and Y=t+1
look_back = 1
print("Size of data sent:",train.shape)
print(len(train)-look_back-1)
trainX, trainY = create_dataset(train, look_back)
testX, testY = create_dataset(test, look_back)
print("TestX",testX.shape)
print(testY.shape)
print(trainX.shape)
#print(trainY)
print(trainY.shape)
#print(trainX.reshape(no_days,1,no_data_per_day*features))
print(no_data_per_day, features)

print(no_data_per_day*features)
trainX = trainX.reshape(725,1,no_data_per_day*features)#9861
trainY = trainY.reshape(725,no_data_per_day*features)
```

```
from keras.layers import Masking
from phased_lstm_keras.PhasedLSTM import PhasedLSTM as PLSTM
a=no_data_per_day*features
print(a)
# create and fit the LSTM network
model = Sequential()
#model.add(Masking(mask_value=3.6122181e+00, batch_input_shape=(None,None, a)))
#(output size,batch_input_shape:input shape of the data)
model.add(PLSTM(a))#, batch_input_shape=(None,None, a),return_sequences=False))
#model.add(TimeDistributed(Dense(5)))
model.add(Dense(a))#,activation='softmax'))
#model.add(Dropout(1))
#model.add(Dense(a))
model.compile(loss='mean_squared_error', optimizer='adam')
#print(model.summary())
model.fit(trainX, trainY, epochs=2, batch_size=30, verbose=2)
#model.fit_generator(generator, steps_per_epoch=1, epochs=500, verbose=0)
```
